```python
import pandas as pd
```

```python
# read in data
data = pd.read_csv(r'C:\Users\Huzaifa\Documents\ML\CrimeData\PathPlanning\interventionscitoyendo.csv', encoding='ISO-8859-1')
```

```python
data.head()
```

```python
# filter data for specified latitude and longitude coordinates and store in 'inSet'
inSet = data[(data.LAT <= 45.512855) & (data.LAT >= 45.502971) & (data.LONG >= -73.583204) & (data.LONG <= -73.570142)]
```

```python
# view filtered data
inSet
```

```python
# filter data for crimes at night within the geographical range we filtered
atNight = inSet[(data.QUART == "nuit")]
```

```python
# view night crimes
atNight
```

```python
# filter data for crimes in evening within the geographical range we filtered
atEvening = atNight = inSet[(data.QUART == "soir")]
```

```python
# view evening crime
atEvening
```

```python
# filter data for crimes during day within the geographical range we filtered

atDay = inSet[(data.QUART == "jour")]
```

```python
# view day crimes
atDay
```
